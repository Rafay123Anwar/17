```python

def wordpatterncheck(pattern,s):
    words=[]
    word=""
    for char in s :
        if char == " " :
            words.append(word)
            word=""
        else :
            word += char
    words.append(word)
    print(words)

    if len(pattern) != len(words) :
        return False

    # print(f"len of pattern : {len(pattern)} ")
    # print(f"len of words : {len(words)} ")

    # print("pattern")
    # for i in range(len(pattern)):
    #     print(pattern[i])

    # print("words")
    # for i in range(len(words)):
    #     print(words[i])



    for i in range(len(pattern)):
        for j in range(i+1,len(pattern)):
            if pattern[i]==pattern[j] and words[i] != words[j] :
                print("this pattern[i]==pattern[j] call")
                return False
            if pattern[i] != pattern[j] and words[i] == words[j] :
                print("this pattern[i] != pattern[j] call")
                return False
    return True


pattern="abcb"
s="dog fish cat fish"

print(wordpatterncheck(pattern,s))
```
