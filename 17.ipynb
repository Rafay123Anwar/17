```python
import pandas as pd
data=pd.read_csv(r"/content/passenger_data.csv")
```

```python
Total_passenger=data['passenger_id'].count() # count function count number of passeger id
print(Total_passenger)
```

```python
passengers_destination=data['destination'].value_counts() # same as count function but value_count count number of value according to destinagation
print(passengers_destination)
```

```python
average_ticket_price=data.groupby('gender')['ticket_price'].mean() # first we apply group function to group according gender and calculate averge ticket price
print(average_ticket_price)
```

```python

def get_passengers_by_destination(dest):                          # this function takes parameters as destination and in this we check destination according to name
    passengers = data[data['destination'] == dest]['name'].tolist()
    return passengers

def get_highest_paying_passenger():                         # in this function  we calculate highest paid passenger by idxmax()
    highest_paying = data.loc[data['ticket_price'].idxmax()]
    return highest_paying['name'], highest_paying['destination']


print(get_passengers_by_destination('Lahore'))
print(get_highest_paying_passenger())
```

```python
sorted_passengers = data.sort_values(by='ticket_price', ascending=False) # sort vales use for sorting the data or column
print(sorted_passengers)

top_3_passengers = sorted_passengers.head(3)      # Head use for access number of row
print(top_3_passengers[['name', 'age', 'ticket_price']])
```

```python

age_threshold = int(input("Enter an age threshold: "))
younger_passengers = data[data['age'] < age_threshold]
print(f"{len(younger_passengers)} passengers are younger than {age_threshold} years old.")
```
